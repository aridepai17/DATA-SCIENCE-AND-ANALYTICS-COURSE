## IMPORTING MAJOR LIBRARIES

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv(r"C:\module5\Iris (1).csv")
iris.head()
```

```python
iris['Species'].value_counts()
```

```python
iris.plot( kind = 'scatter', x = 'SepalLengthCm', y = 'SepalWidthCm')
plt.show()
```

```python
sns.jointplot(x = 'SepalLengthCm', y = 'SepalWidthCm', data = iris, size = 5)
plt.show()
```

```python
sns.FacetGrid(iris, hue = 'Species').map(plt.scatter, 'SepalLengthCm', 'SepalWidthCm').add_legend()
plt.show()
```

```python
sns.boxplot(x = 'Species', y = 'PetalLengthCm', data = iris, palette = 'colorblind')
plt.show()
```

```python
ax = sns.boxplot( x = 'Species', y = 'PetalLengthCm', data = iris)
ax = sns.stripplot(x = 'Species', y = 'PetalLengthCm', data = iris, jitter = True, edgecolor = 'gray')
plt.show()
```

```python
sns.violinplot(x = 'Species', y = 'PetalLengthCm', data = iris, palette = 'colorblind')
plt.show()
```

```python
sns.FacetGrid(iris, hue = 'Species')\
.map(sns.kdeplot, 'PetalLengthCm')\
.add_legend()
plt.show()
```

```python
sns.pairplot(iris.drop('Id', axis = 1), hue = 'Species', size = 3)
plt.show()
```

```python
sns.pairplot(iris.drop("Id", axis= 1), hue = 'Species', size = 3, diag_kind = 'box')
plt.show()
```

```python
iris.drop("Id", axis = 1).boxplot(by = "Species", figsize = (12, 16))
plt.show()
```

```python
from pandas.plotting import andrews_curves

andrews_curves(iris.drop("Id", axis = 1), "Species")
```

```python
from pandas.plotting import parallel_coordinates

parallel_coordinates(iris.drop("Id", axis = 1), "Species")
plt.show()
```

```python
from pandas.plotting import radviz
radviz(iris.drop("Id", axis = 1), "Species")
plt.show()
```
