## STANDARDIZATION AND NORMALIZATION OF DATA

## IMPORTING MAJOR LIBRARIES

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
wine_df = pd.read_csv(r"C:\winedataset\winequality-red.csv")
```

```python
wine_df
```

```python
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler
import matplotlib.pyplot as plt
```

```python
mmscaler = MinMaxScaler()
minmax = mmscaler.fit_transform(wine_df)
minmax = pd.DataFrame(minmax, columns = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol', 'quality'])
minmax
```

```python
sscaler = StandardScaler()
standard = sscaler.fit_transform(wine_df)
standard = pd.DataFrame(standard, columns = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol', 'quality'])
standard
```

```python
rscaler = RobustScaler()
robust = rscaler.fit_transform(wine_df)
robust = pd.DataFrame(robust, columns = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol', 'quality'])
robust
```

```python
plt.figure(figsize = (15,4))
plt.subplot(1, 4, 1)
sns.kdeplot(wine_df['fixed acidity'], color = 'red')
sns.kdeplot(wine_df['citric acid'], color = 'green')
sns.kdeplot(wine_df['total sulfur dioxide'], color = 'blue')
sns.kdeplot(wine_df['alcohol'], color = 'brown')
plt.title('Without Scaling the data')
plt.show()
```

```python
plt.figure(figsize = (15,4))
plt.subplot(1, 4, 1)
sns.kdeplot(minmax['fixed acidity'], color = 'red')
sns.kdeplot(minmax['citric acid'], color = 'green')
sns.kdeplot(minmax['total sulfur dioxide'], color = 'blue')
sns.kdeplot(minmax['alcohol'], color = 'brown')
plt.title('Using Min-Max Scaler on the data')
plt.show()
```

```python
plt.figure(figsize = (15,4))
plt.subplot(1, 4, 1)
sns.kdeplot(standard['fixed acidity'], color = 'red')
sns.kdeplot(standard['citric acid'], color = 'green')
sns.kdeplot(standard['total sulfur dioxide'], color = 'blue')
sns.kdeplot(standard['alcohol'], color = 'brown')
plt.title('Using Standard Scaler on the data')
plt.show()
```

```python
plt.figure(figsize = (15,4))
plt.subplot(1, 4, 1)
sns.kdeplot(robust['fixed acidity'], color = 'red')
sns.kdeplot(robust['citric acid'], color = 'green')
sns.kdeplot(robust['total sulfur dioxide'], color = 'blue')
sns.kdeplot(robust['alcohol'], color = 'brown')
plt.title('Using Robust Scaler on the data')
plt.show()
```

## DISTRIBUTION OF DATA WITH SCALERS

```python
#Without Scaling
plt.figure(figsize = ( 14, 8))
for i, col in enumerate(wine_df.columns, 1):
    plt.title(f'Distribution of {col} Data')
    sns.histplot(wine_df[col], kde = True)
    plt.tight_layout()
    plt.plot()
    plt.show()
```

```python
# Min Max Scaler

plt.figure(figsize = ( 14, 8))
for i, col in enumerate(minmax.columns, 1):
    plt.title(f'Distribution of {col} Data')
    sns.histplot(minmax[col], kde=True)
    plt.tight_layout()
    plt.plot()
    plt.show()
```

```python
# Standard Scaler

plt.figure(figsize = ( 14, 8))
for i, col in enumerate(standard.columns, 1):
    plt.title(f'Distribution of {col} Data')
    sns.histplot(standard[col], kde=True)
    plt.tight_layout()
    plt.plot()
    plt.show()
```

```python
# Robust Scaler

plt.figure(figsize = ( 14, 8))
for i, col in enumerate(robust.columns, 1):
    plt.title(f'Distribution of {col} Data')
    sns.histplot(robust[col], kde=True)
    plt.tight_layout()
    plt.plot()
    plt.show()
```
