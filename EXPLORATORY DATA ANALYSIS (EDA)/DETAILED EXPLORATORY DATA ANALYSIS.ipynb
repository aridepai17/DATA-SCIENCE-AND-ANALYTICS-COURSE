## OUTLIER DETECTION AND REMOVAL

```python
numerical = [ 'age', 'weight', 'height', 'bmi']
data[numerical].describe()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 10)) 

sns.histplot(data=data, x='age', kde=True, color='red', ax=axs[0, 0])
sns.histplot(data=data, x='height', kde=True, color='skyblue', ax=axs[0, 1])
sns.histplot(data=data, x='weight', kde=True, color='olive', ax=axs[1, 0])
sns.histplot(data=data, x='bmi', kde=True, color='gold', ax=axs[1, 1])

plt.tight_layout() 
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 18)) 

sns.boxplot(data['age'], ax=axs[0, 0])
sns.boxplot(data['height'], ax=axs[0, 1])
sns.boxplot(data['weight'], ax=axs[1, 0])
sns.boxplot(data['bmi'], ax=axs[1, 1])

plt.tight_layout()
plt.show()
```

```python
qh25, qh50, qh75 = np.percentile(data['height'], [ 25, 50, 75])

iqr = qh75 - qh25

minh = qh25 - 1.5 * iqr
maxh = qh75 + 1.5 * iqr
```

```python
better = data[(data['height'] > minh) & (data['height'] < maxh)]
better
```

```python
qb25, qb50, qb75 = np.percentile(data['bmi'], [25,50,75])

iqrb = qb75 - qb25

maxb = qb75 + 1.5 * iqrb

best = better[better['bmi'] < maxb]
best
```

```python
best.shape
```

```python
fig, axs = plt.subplots(1, 3, figsize = (15, 5))

sns.histplot(data = data, x = 'weight', kde = True, color = 'skyblue', ax = axs[0])
sns.histplot(data = data, x = 'height', kde = True, color = 'skyblue', ax = axs[1])
sns.histplot(data = data, x = 'bmi', kde = True, color = 'gold', ax = axs[2])

plt.show()
```
