## RENAMING THE COLUMNS

```python
df = df.rename( columns = {"Engine HP": "HP", "Engine Cylinders":"Cylinders", "Transmission Type": "Transmission", "Driven_Wheels": "Drive Model","Highway":"MPG-H", "city mpg": "MPG-C", "MSRP": "Price"})
df
```

## DROPPING DUPLICATE ROWS

```python
df.shape
```

```python
duplicaterows = df[df.duplicated()]
print("Number of duplicate rows: ", duplicaterows.shape)
```

```python
df = df.drop_duplicates()
df
```

## DROP THE MISSING AND NULL VALUES

```python
df.isnull().sum()
```

```python
df = df.dropna()
df.count()
```

```python
df.isnull().sum()
```

## DETECTING OUTLIERS

```python
sns.boxplot(x = df['Price'])
plt.show()
```

```python
sns.boxplot(x = df['HP'])
plt.show()
```

```python
# Step 1: Select only numeric columns
df_numeric = df.select_dtypes(include='number')

# Step 2: Compute IQR
Q1 = df_numeric.quantile(0.25)
Q3 = df_numeric.quantile(0.75)
IQR = Q3 - Q1

# Step 3: Create a mask to identify non-outlier rows
mask = ~((df_numeric < (Q1 - 1.5 * IQR)) | (df_numeric > (Q3 + 1.5 * IQR))).any(axis=1)

# Step 4: Apply the mask to the full original DataFrame
df = df[mask]
df
```

```python
df.shape
```

## PLOT DIFFERENT FEATURES AGAINST ONE ANOTHER USING SCATTER PLOT, HISTOGRAM AND HEAT MAP

```python
# Histogram

df.Make.value_counts().nlargest(40).plot(kind = 'bar', figsize = (10, 5))
plt.title('Number of Cars by Make')
plt.ylabel('Number of Cars')
plt.xlabel('Make')
plt.show()
```

```python
## Scatter Plot

fig, ax = plt.subplots(figsize = ( 10, 6))
ax.scatter(df['HP'], df['Price'])
ax.set_xlabel('HP')
ax.set_ylabel('Price')
plt.show()
```

```python
# Heat Map
df_numeric = df.select_dtypes(include='number')

# Step 2: Compute correlation matrix
corr_mat = df_numeric.corr()

# Step 3: Plot the heatmap
plt.figure(figsize=(10, 10))
sns.heatmap(corr_mat, cmap='coolwarm', annot=True, fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()
```
