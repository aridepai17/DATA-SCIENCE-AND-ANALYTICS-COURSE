## LINEAR REGRESSION

## IMPORTING MAJOR LIBRARIES

```python
import pandas as pd
import sklearn
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
```

```python
test_lr = {
    'A' : [1, 2, 3, 4, 5],
    'B' : [2, 4, 6, 8, 10]
}
```

```python
test_lr_df = pd.DataFrame(test_lr)
```

```python
Feature = pd.DataFrame(test_lr_df['A'])
```

```python
target = pd.DataFrame(test_lr_df['B'])
```

```python
Lr = LinearRegression()
```

```python
Lr.fit(Feature, target)
```

```python
LinearRegression()
```

```python
Lr.coef_
```

```python
Lr.intercept_
```

```python
# y = 2x + 0
```

```python
xvalues = np.linspace(min(Feature.values), max(Feature.values), 100)
yvalues = Lr.coef_ * xvalues + Lr.intercept_
```

```python
plt.scatter(Feature, target, label = 'data')
plt.plot(xvalues, yvalues)
plt.xlabel('Independent Variable : A')
plt.ylabel('Dependent Variable : B')
plt.title('Linear Regresssion')
plt.show()
```

## PERFORMING LINEAR REGRESSION ON THE SECOND HAND PRICE PREDICTION DATASET

```python
carsdf = pd.read_csv(r"C:\module 7\cars.csv")
carsdf
```

```python
Features = pd.DataFrame(carsdf['Kilometers_Driven'])
```

```python
targets = pd.DataFrame(carsdf['Price'])
```

```python
Lr2 = LinearRegression()
```

```python
Lr2.fit(Features, targets)
```

```python
Lr2.coef_
```

```python
Lr2.intercept_
```

```python
# y = -5.6x + 1731756.88
```

```python
x_values = np.linspace(min(Features.values), max(Features.values), 100)
y_values = Lr2.coef_ * x_values + Lr2.intercept_
```

```python
plt.scatter(Features, targets, label = "Second Hand Cars Data")
plt.plot(x_values, y_values)
plt.title('Linear Regression on Second Hand Cars Dataset')
plt.show()
```

## PERFORMING TRAIN TEST SPLIT

```python
x_train, x_test, y_train, y_test = train_test_split(Features, targets, test_size = 0.2, random_state = 1)
```

```python
Lr2.fit(x_train, y_train)
```

```python
Lr2.coef_
```

```python
Lr2.intercept_
```

```python
y_pred = Lr2.predict(x_test)
y_pred
```

```python
mean_absolute_error(y_test, y_pred)
```
