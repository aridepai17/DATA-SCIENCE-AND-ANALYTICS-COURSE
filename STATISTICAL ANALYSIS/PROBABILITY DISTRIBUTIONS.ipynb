## IMPORTING MAJOR LIBRARIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## CONTINUOUS DISTRIBUTION

### NORMAL DISTRIBUTION

```python
mu, sigma = 0, 0.1
s = np.random.normal(mu, sigma, size = 1000 )
count, bins, ignored = plt.hist(s, 30, density = True)
plt.title('Normal Distribution')
plt.show()
```

### UNIFORM DISTRIBUTION

### CAUCHY DISTRIBUTION

```python
c = np.random.standard_cauchy(100000)
c = c[(c>-25) & (c<25)]
plt.hist(c, bins = 100, density = True)
plt.title('Cauchy Distribution')
plt.show()
```

### T DISTRIBUTION

```python
t = np.random.standard_t(df = 1, size = 100000)
t = t[np.where(abs(t) < 5)]
h = plt.hist(t, bins = 100, density = True)
plt.title('t Distribution')
plt.show()
```

### F DISTRIBUTION

```python
f = np.random.f(dfnum = 1, dfden = 1, size = 10000)
f = f[np.where(abs(f) < 5)]
h = plt.hist(f, bins = 100, density = True)
plt.title('f Distribution')
plt.show()
```

### CHI SQUARE DISTRIBUTION

```python
q = np.random.chisquare(df = 1, size = 10000)
q = q[np.where(abs(q) < 20 )]
h = plt.hist(q, bins = 100, density = True)
plt.title('Chi Square Distributon')
plt.show()
```

### EXPONENTIAL DISTRIBUTION

```python
e = np.random.exponential(scale = 1.0, size = 100000)
h = plt.hist(e, bins = 1000, density = True)
plt.title('Exponential Distribution')
plt.show()
```

### WEIBULL DISTRIBUTION

```python
w = np.random.weibull( a = 0.5, size = 10000)
w = w[np.where(abs(w) < 5)]
h = plt.hist(w, bins = 100, density = True)
plt.title('Weibull Distribution')
plt.show()
```

### LOG NORMAL DISTRIBUTION

```python
l = np.random.lognormal(mean = 0.0, sigma = 0.5, size = 10000)
l = l[np.where(abs(l) < 5)]
h = plt.hist(l, bins = 100, density = True)
plt.title('Log Normal Distribution')
plt.show()
```

### GAMMA DISTRIBUTION

```python
g = np.random.gamma( shape = 0.5, scale = 1.0, size = 10000)
g = g[np.where(abs(g) < 10)]
h = plt.hist(g, bins = 100, density = True)
plt.title('Gamma Distribution')
plt.show()
```
